```python
# To get the Last bill no

def last_bill_no():
    cursor.execute('select max(bill_id) from bills')
    record = cursor.fetchone()
    return record
```

```python
# To get sale by datetime
def date_wise_sell():
    clear()

    print('Sell Between Two Dates -- Screen')
    print('-'*100)
    start_date = input('Enter start Date (YYYY-MM-DD) :')
    end_date   = input('Enter End Date (YYYY-MM-DD) :')

    sql = f'select * from bills where bill_date between "{start_date}" and "{end_date}"'

    cursor.execute(sql)
    records = cursor.fetchall()

    clear()

    print('Bill No\t Customer Name\t\t Phone No    \t Bill Date')

    print('-'*100)

    for row in records:
        print(f'{row[0]:<10d} {row[1]:20s} {row[2]:20s} {row[3]}')

    print('-'*100)

    print('\n\nPress any key to continue....')

    wait= input()
```

```python
def bill_information():
    clear()
    bill_no = input('Enter Bill Number :')

    sql = F"""select b.bill_id, b.name, b.phone, b.bill_date, t.item_id,t.qty,i.item_name,
       i.price from bills b,transactions t,items i 
       where b.bill_id = t.bill_id AND t.item_id= i.id AND 
       b.bill_id ={bill_no};"""


    cursor.execute(sql) 

    records = cursor.fetchall()

    n = cursor.rowcount

    clear()

    print("Bill No :",bill_no)
    print('-'*100)

    if n<=0:
        print(f'Bill number {bill_no} does not exists')

    else:
        print(f'Customer Name : {records[0][1]}  Phone No :{records[0][2]}')

        print(f'Bill Date : {records[0][3]}'.format(records[0][3]))

        print('-'*100)

        print('{:10s} {:30s} {:20s} {:10s} {:30s}'.format('ID','Item Name','Qty',
             'Price','Amount'))
        print('-'*100)

        for row in records:
            print('{:<10d} {:30s} {:<20d} {:.2f}    {:>.2f}'.format(row[4],row[6],row[5],row[7],row[5]*row[7]))

        print('-'*100)

    print('\nPress any key to continue....')
    wait = input()
```

```python
# To get overall collection as per date 
def amount_collected():
    clear()
    
    start_date = input('Enter start Date (YYYY-MM-DD) :')
    end_date   = input('Enter End   Date (YYYY-MM-DD) :')
    clear()

    print(f'Amount collected between: {start_date} and {end_date}')
    print('-'*100)

    sql = f'select round(sum(t.qty*i.price),2) from bills b,transactions t,items i \
           where b.bill_date between "{start_date}" AND "{end_date}" AND b.bill_id = t.bill_id AND \
           t.item_id = i.id;'
    cursor.execute(sql)
    result = cursor.fetchone()[0]

    print(result)

    print('\nPress any key to continue.....')
    wait= input()
```
